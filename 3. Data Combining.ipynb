```python
import pandas as pd
```

```python
df1 = pd.read_csv('books.csv')
```

```python
df2 = pd.read_csv('data.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.DataFrame()

df['title']         = df1['title']
df['upc']           = df2['upc']
df['category']      = df2['category']
df['price_exc_tax'] = df2['price_exc_tax']
df['price_inc_tax'] = df2['price_inc_tax']
df['tax']           = df2['tax']
df['rating']        = df1['rating']
df['reviews']       = df2['reviews']
df['quantity']      = df2['quantity']
df['stock']         = df1['status']

df['book_link']     = df1['book_link']
df['img_link']      = df1['img_link']
```
