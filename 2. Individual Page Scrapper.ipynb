```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
df = pd.read_csv('books.csv')
```

```python
df
```

### Page Scraping

```python
data = []
for link in tqdm(df['book_link']):
    
    res  = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    book_type     = soup.find_all('a')[3].text   
    upc           = soup.find_all('td')[0].text
    price_x       = soup.find_all('td')[2].text[2:]
    price_i       = soup.find_all('td')[3].text[2:]
    tax           = soup.find_all('td')[4].text[2:]
    availability  = soup.find_all('td')[5].text
    no_of_reviews = soup.find_all('td')[6].text
    
    data.append([book_type, price_x, price_i, tax, availability, upc, no_of_reviews])
```

```python
df = pd.DataFrame(data, columns = ['category', 'price_exc_tax', 'price_inc_tax', 'tax', 'quantity', 'upc', 'reviews'])
```
